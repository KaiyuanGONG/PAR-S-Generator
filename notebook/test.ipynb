把case_0000_atn_av.bin这个二进制文件 读出来打印一下平均值
- 如果都是很大的数，你需要把它映射/缩放（Rescale）到水=0.15 左右的衰减系数值

```python
import numpy as np

path = "case_0000_atn_av.bin"

x = np.fromfile(path, dtype=np.float32)
print("count:", x.size)
print("mean:", float(x.mean()))
```

```python
import numpy as np
import os

path = "case_0000_atn_av.bin"

size = os.path.getsize(path)
print("file size:", size, "bytes")

if size % 4 == 0:
    print("Possible dtype: float32")
    x32 = np.fromfile(path, dtype=np.float32)
    print("float32 mean:", float(x32.mean()))

if size % 8 == 0:
    print("Possible dtype: float64")
    x64 = np.fromfile(path, dtype=np.float64)
    print("float64 mean:", float(x64.mean()))
```
